```python
import warnings
warnings.filterwarnings(action='ignore',category=FutureWarning)
warnings.filterwarnings(action='ignore',category=RuntimeWarning)
import pandas as pd
%load_ext rpy2.ipython
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
from matplotlib import cm, pyplot as plt
%matplotlib inline
```

```python
temp_data = pd.read_csv('FinalTempdata1.csv')
denver_final = pd.read_csv('FinalTempYr3.csv')
energy_data = pd.read_csv('EnergyData6.csv')
```

```python
city_map = temp_data.City
energy_data.columns=temp_data.drop('City',axis=1).columns
def GetCityName(city_index):
    return city_map[city_index]
def logTransform(x):
    return np.log(x+1)
def logTransformInverse(x):
    return np.exp(x)-1
assert(logTransformInverse(logTransform(42)) == 42), "Something unexpected happened in my transforms"
```

```python
allData=temp_data.drop('City',axis=1).stack().reset_index()
allData.columns=['City','MoYr','Temperature']
allData['Energy']=energy_data.stack().reset_index()[0]
allData.City = allData.City.apply(GetCityName)
allData['Year']=allData.MoYr.apply(lambda x: int(x[-1:]))
allData['Month']=allData.MoYr.apply(lambda x: x[:3])
allData=allData.drop('MoYr',axis=1)
allData=allData[['City','Year','Month','Temperature','Energy']]
```

```python
allData['LogEnergy'] = allData.Energy.apply(logTransform)
denver_data = allData[allData.City == 'Denver, Colorado']
denver_final['LogEnergy'] = 0
denver_final['Temperature']=denver_final.Temp
denver_final=denver_final.drop('Temp',axis=1)
denver_final['City']='Denver, Colorado'
```

```python
month_color={}
for j,i in enumerate(allData.Month.unique()):
    month_color[i] = cm.viridis(j/12)
year_cm = cm.Set1
year_color={1:year_cm(0),2:year_cm(1)}
```

```python
months=allData.Month.unique().tolist()
def GetMonthNo(month_name):
    return months.index(month_name)+1
allData['MonthNo'] = allData.Month.apply(GetMonthNo)
```

```python
fig, (ax1,ax2) = plt.subplots(2,2,figsize=(10,10),sharey='row',dpi=200)
cGroup = allData.groupby("City")
for year,ygroup in allData.groupby("Year"):
    for city,cgroup in ygroup.groupby("City"):
        ax1[year-1].plot(cgroup.Month,cgroup.Temperature,label='_nolegend_')
        ax2[year-1].plot(cgroup.Month,cgroup.Energy,label='_nolegend_')
    # Replot Denver annotated and in bold
    ygroup_denver = ygroup[ygroup.City == 'Denver, Colorado']
    ax1[year-1].plot(ygroup_denver.Month,ygroup_denver.Temperature,'k-',linewidth=5,label='Denver')
    ax2[year-1].plot(ygroup_denver.Month,ygroup_denver.Energy,'k-',linewidth=5,label='Denver')
[ax.legend(loc=1) for ax in ax1]
[ax.legend(loc=1) for ax in ax2]
[ax.grid(True,alpha=0.3) for ax in ax1]
[ax.grid(True,alpha=0.3) for ax in ax2]
ax1[0].set_ylabel("Temperature ($\degree$ C)")
ax2[0].set_ylabel("Energy usage (hundreds GWh)")
fig.subplots_adjust(wspace=0)
fig.canvas.draw()
fig.savefig('exploratory1.png',dpi=fig.dpi,transparent=True)
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,6),dpi=200)
for key,group in allData.groupby('Year'):
    group.plot.scatter(x='Temperature',y='Energy',ax=ax[0],color=year_color[key],label='Year '+str(key))
ax[0].set_ylabel("Energy usage (hundreds GWh)",fontsize=16)
ax[0].set_title("Energy usage vs Temperature",fontsize=16)
ax[0].legend(fontsize=16)
grouped = allData.groupby('Year',sort=False)
for key, group in grouped:
        group.plot(ax=ax[1],kind='scatter',x='Temperature', y='LogEnergy', label='Year '+str(key),color=year_color[key])
ax[1].set_ylabel('g(Energy usage)',fontsize=16)
[a.set_xlabel('Temperature ($\degree$C)',fontsize=16) for a in ax]
ax[1].set_title("g(Energy usage) vs Temperature",fontsize=16)
ax[1].legend(fontsize=16)
[x.grid(True,alpha=0.3) for x in ax]
fig.canvas.draw()
fig.savefig('exploratory2.png',dpi=fig.dpi,transparent=True)
```

```python
cities=['Denver, Colorado','Miami, Florida','New York, New York','San Francisco, California']
fig,ax=plt.subplots(1,1,figsize=(7,7),dpi=200)
city_color={}
for j,i in enumerate(cities):
    city_color[i] = cm.Set1(j)
for city in cities:
    city_data = allData[allData.City == city]

    model=sm.OLS(city_data.LogEnergy,sm.add_constant(city_data.Temperature))
    mfit = model.fit()

    T=np.linspace(city_data.Temperature.min(),city_data.Temperature.max(),100)
    pred=mfit.get_prediction(sm.add_constant(T))
    ci=pred.conf_int(alpha=0.05).transpose()
    pm=pred.predicted_mean

    ax.scatter(city_data.Temperature,city_data.LogEnergy,color=city_color[city],label=city)
    ax.plot(T,pm,color=city_color[city],linestyle='dashed',label=None)
ax.grid(True)
ax.legend(fontsize=13)
ax.set_title("Energy usage trends for select cities",fontsize=20)
ax.set_ylabel("g(Energy usage)",fontsize=18)
ax.set_xlabel("Temperature ($\degree$C)",fontsize=18)
fig.canvas.draw()
fig.savefig('select_cities.png',dpi=fig.dpi,transparent=True)
```

```python
%%R
reModel <- lmer(LogEnergy ~ Temperature + (1 | City) + (0 + Temperature|City), data=allData)

myPrediction <- function(.) {
    fit <- predict(reModel, . ,interval='prediction')
    mm <- model.matrix (terms(reModel), .)
    pvar1 <- diag(mm %*% tcrossprod(vcov(reModel),mm))
    tvar1 <- pvar1+VarCorr(reModel)$City[1]
    cmult <- 1.96
    pred <- data.frame(fit,
                       lwr = fit - cmult*sqrt(tvar1),
                       upr = fit + cmult*sqrt(tvar1))
    return(pred)
    }

print(summary(reModel))
png('lmer_diagnostic%03d.png',width=1024,height=1024,res=200,bg=NA)
plot(lme4:::plot.merMod(reModel,xlab='Fitted values',ylab='Pearson residuals'))
dev.off()
```

```python
%%R
overdisp_fun <- function(model) {
  ## number of variance parameters in 
  ##   an n-by-n variance-covariance matrix
  vpars <- function(m) {
    nrow(m)*(nrow(m)+1)/2
  }
  model.df <- sum(sapply(VarCorr(model),vpars))+length(fixef(model))
  rdf <- nrow(model.frame(model))-model.df
  rp <- residuals(model,type="pearson")
  Pearson.chisq <- sum(rp^2)
  prat <- Pearson.chisq/rdf
  pval <- pchisq(Pearson.chisq, df=rdf, lower.tail=FALSE)
  c(chisq=Pearson.chisq,ratio=prat,rdf=rdf,p=pval)
}
```
